## Models

```python
X = df.drop(['Churn'], axis=1)
y = df[['Churn']]
columns = X.columns
```

### Handling unbalanced data

#### Over sampling

To handle the unbalance of our data, we will try oversampling with SMOTE technique.

```python
smote = SMOTE(sampling_strategy=1, random_state=777, n_jobs=-1)
```

```python
X=np.array(X)
X, y = smote.fit_resample(X, y.values.ravel())
```

```python
X = pd.DataFrame(X, columns=columns)
```

```python
# 10% of data is sufficient to test our model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=777) 
```

```python
churn_rate()
```

### Logistic regression

```python
log_reg = LogisticRegression(max_iter=10000, random_state=777)
log_reg.fit(X_train,y_train)
y_pred = log_reg.predict(X_test)
y_pred_train = log_reg.predict(X_train)
```

#### Feature importances

```python
colors=['#FFAEBC','#A0E7E5', '#B4F8C8', '#FBE7C6', '#79BEEE']

feature_importance = pd.DataFrame(log_reg.feature_names_in_, columns = ["Feature"])
feature_importance["importance"] = pow(math.e, log_reg.coef_[0])
feature_importance = feature_importance.sort_values(by = ["importance"], ascending=False)

ax = feature_importance.plot(x='Feature', y='importance', kind='bar', color = colors,figsize = (15,8))
ax.set_ylabel('Importance',size = 14)
ax.set_xlabel('Feature',size = 14)
ax.set_title('Feature importances', size = 14)
```

```python
print('Accuracy for the test dataset: {}'.format(accuracy_score(y_test, y_pred)))
print('Accuracy for the train dataset: {}'.format(accuracy_score(y_train, y_pred_train)))
```

#### Confusion matrix

```python
cm = confusion_matrix(y_test, y_pred, labels=log_reg.classes_, normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=log_reg.classes_)
disp.plot()
plt.show()
```

```python
print("Precision score: ", precision_score(y_test, y_pred))
print("Recall score: ", recall_score(y_test, y_pred))
```

As expected, accuracy alone doesn't give a clear idea of the performance of our model. We have to see precision and recall.

A better way to test our model is to use Scikit-Learn's K-fold cross-validation feature. The following code randomly splits the training set into 10 distinct subsets called folds, then it trains and evaluates our model 10 times, picking a different fold for evaluation every time and training on the other 9 folds. The result is an array containing the 10 evaluation scores.

```python
metrics = ['accuracy', 'precision', 'recall', 'f1']
    
# scores = cross_val_score(log_reg, X, y, cv=10, n_jobs=-1, scoring=metrics)

scores = cross_validate(log_reg, X, y, cv=10, n_jobs=-1, scoring=metrics)

def display_scores(scores):
    # Accuracy
    print("----------------------------------------Accuracy-----------------------------------------")
    print("Accuracy scores: ", scores['test_accuracy'])
    print("Accuracy mean: ", scores['test_accuracy'].mean())
    print("Accuracy standard deviation: ", scores['test_accuracy'].std())
    
    # Precision
    print("\n--------------------------------------Precision----------------------------------------")
    print("Precision scores: ", scores['test_precision'])
    print("Precision mean: ", scores['test_precision'].mean())
    print("Precision standard deviation: ", scores['test_precision'].std())
    
    # Recall
    print("\n---------------------------------------Recall------------------------------------------")
    print("Recall scores: ", scores['test_recall'])
    print("Recall mean: ", scores['test_recall'].mean())
    print("Recall standard deviation: ", scores['test_recall'].std())
    
    # F1
    print("\n-----------------------------------------F1--------------------------------------------")
    print("F1 scores: ", scores['test_f1'])
    print("F1 mean: ", scores['test_f1'].mean())
    print("F1 standard deviation: ", scores['test_f1'].std())

display_scores(scores)
```

### XGBoost

```python
model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy: ",accuracy_score(y_test, y_pred))
print("ROC: ",roc_auc_score(y_test, y_pred))
print("Precision: ",precision_score(y_test, y_pred))
print("Recall: ",recall_score(y_test, y_pred))
```
